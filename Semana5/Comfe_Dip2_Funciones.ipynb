```python
def devolver_distintos(a,b,c):
  suma=a+b+c
  lista=[a,b,c]
  #      0 1 2
  if suma>15:
    return max(lista)
  elif suma<10:
    return min(lista)
  else:
    lista.sort()
    return lista[1]

n1=int(input("Ingrese primer numero de la lista"))
n2=int(input("Ingrese segundo numero de la lista"))
n3=int(input("Ingrese tercer numero de la lista"))

valor_devuelto=devolver_distintos(n1,n2,n3)
print(valor_devuelto)

```

```python
def letras_unicas(palabra):
  conjunto=set()
  for letra in palabra:
    conjunto.add(letra)
  
  mi_lista=list(conjunto)
  mi_lista.sort()
  return mi_lista

palabra=input('Ingrese Palabra')
print(letras_unicas(palabra))
```

```python
def ceros_vecinos(lista):
  contador=0
  print(lista)
  for numero in lista:
    if lista[contador]==0 and lista[contador+1]==0:
      return True
    else:
      contador+=1
  return False

lista=[]
n=int(input('Cuantos elementos desea ingresar a la lista'))
for i in range(n):
  numero=int(input('Ingrese numero'))
  lista.append(numero)

print(ceros_vecinos(lista))

```
